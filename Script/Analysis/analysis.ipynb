```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
paths = [
        # 'Tests\case-example_OUTPUT.txt',
        '..\..\Tests\caseG50_OUTPUT.txt',
        '..\..\Tests\caseG100_OUTPUT.txt',
        '..\..\Tests\caseG200_OUTPUT.txt',
        '..\..\Tests\caseG500_OUTPUT.txt',
        '..\..\Tests\caseG750_OUTPUT.txt',
        '..\..\Tests\caseG1000_OUTPUT.txt',
        '..\..\Tests\caseG1500_OUTPUT.txt',
        '..\..\Tests\caseG2000_OUTPUT.txt'
    ]
```

```python
# Tamanho dos dados
x = [50, 100, 200, 500, 750, 1000, 1500, 2000]

operations = []
for path in paths:
   with open(path, 'r') as file:
        operations.append(int(file.readline().split()[-1]))
print(operations)


# Funções de complexidade
y_n = np.array(x)
y_n_squared = np.array([n**2 for n in x])
y_n_cubed = np.array([n**3 for n in x])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y_n, label='O(N)')
plt.plot(x, y_n_squared, label='O(N^2)')
plt.plot(x, y_n_cubed, label='O(N^3)')
plt.plot(x, operations, label='Contagem de Operações', marker='o', linestyle='--', color='black')

plt.title('Complexidade de Tempo vs Contagem de Operações')
plt.xlabel('Tamanho dos Dados')
plt.ylabel('Complexidade de Tempo / Contagem de Operações')
plt.legend()
plt.grid(True)
plt.xticks(x, rotation=45)
plt.tight_layout()

plt.savefig('complexity.png')
plt.show()
```
